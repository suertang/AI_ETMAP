```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

#此函数接受一个参数并返回轨压列表
def genrailp(count):    
    #如果一共设定了n次轨压，最高轨压即为n*100bar，其他轨压顺延，最后一个轨压是250bar
    r=[i*100000 for i in range(count+1,2,-1)]
    r+=[250000]
    return r
def railpgen(se,cnt):
    railp_=genrailp(cnt)
    match=0
    for k,v in se.iteritems():
        if(v==1):            
            se[k]=railp_[match]
            match+=1
m=pd.read_csv("C:/Users/taz3wx/Desktop/Coding/WEB_ET_MAP/20181018143931_CR_Test18_120_ET_map_Yunnei_D30.txt",sep=';',header=None,encoding='latin_1',skiprows=3,skipinitialspace=True)
m.columns=m.loc[0,:]


if("p_Rail(ASAP)" in m.columns):
    print("轨压设定值存在")
else:
    print("轨压设定值丢了")
dd=m.iloc[2:,:]
railp_col=dd.Lastpunkt#.copy()
cnt=dd.Lastpunkt[dd.Lastpunkt=="Set_pRail"].count()
railp_col=(railp_col=="Set_pRail").astype(int)
railp_col.replace(0,np.nan,inplace=True)
railpgen(railp_col,cnt)

Qlist=["Q_1(Emi2)","V_Inj1(HDA_1)"]
head=m.loc[0,:]
Qlabel=head[head.isin(Qlist)][0]
xd=dd.loc[:,["SendAsapCmd(ASAP)_1",Qlabel]]

xd['p_Rail(ASAP)']=railp_col.fillna(method='ffill')
 
#xd['p_Rail(ASAP)'].fillna(method='ffill',inplace=True) 
xd.dropna(subset=[Qlabel],inplace=True) 
xd=xd.apply(pd.to_numeric) 
ff=pd.pivot_table(xd,index='SendAsapCmd(ASAP)_1',columns='p_Rail(ASAP)') 
ff.columns=ff.columns.droplevel()
#p.value_counts()
ff.plot()
```
