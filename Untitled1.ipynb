```python
def func(nums,k):
    for i in range(len(nums)):
        for j in range(i,len(nums)):
            if(nums[i]==nums[j] and abs(i-j)>=k):
                return True
    return False
print(func([1,2,3,1],3))
```

```python
def revint(num):
    if(num==0):
        return 0
    elif(num>0):
        return int("".join(list(str(num))[::-1]))
    else:
        return int("-"+"".join(list(str(num))[1:][::-1]))
revint(-1230)
                   
```

```python
import numpy as np

import matplotlib.pyplot as plt
pts=np.arange(-2,2,0.01)
xs,ys=np.meshgrid(pts,pts)
for i in range(1,10):
    z=np.exp(-(xs**i/2)+(ys**i/2))
    plt.imshow(z,cmap=plt.cm.seismic);plt.colorbar()
    plt.title('hhh={}'.format(i))
    plt.show()
```
