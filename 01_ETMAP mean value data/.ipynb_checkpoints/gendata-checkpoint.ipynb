```python
import pandas as pd
%matplotlib inline
import numpy as np
def getdata(mid):
    df1=pd.read_csv('{0:03d}.csv'.format(int(mid)),skiprows=1,index_col=1)
    df1.dropna(axis=1,inplace=True)
    a=df1.mean()
    df1.index.name="ET"
    #print(df1)
    df2=pd.DataFrame(df1.stack())
    #df2['ET2']=df2.index.droplevel(1)
    #df2['Railp']=df2.index.droplevel()
    df2.reset_index(level=0, inplace=True)
    df2.reset_index(level=0, inplace=True)
    df2.columns=['RailP','ET','InjQ']
    #df2['ID']=float(id)

    df2.index=[mid] * len(df2.index)
    df2.index.name="ID"
    #df2=df2.loc[(df2.ET <= 200) | ((df2.ET % 30==0) & (df2.ET<=500) & (df2.ET>200)) | ((df2.ET % 150==0) & (df2.ET<=2500) & (df2.ET>500))]
    df2=df2.loc[df2.ET<=1500]

    #f2['meq']=df2.RailP.map(lambda x:a[x])
    df2=df2[['RailP','ET','InjQ']]
    return df2


dfe=pd.read_csv('../ET_MAP.csv',skip_blank_lines=True)#这是总结的表格
dfe.dropna(axis=1,how='all',inplace=True)
dfe.dropna(axis=0,how='all',inplace=True)
#dfe=dfe[~dfe['Injector  Type'].str.startswith('CRIN')]
#dfe=dfe[~dfe['Injector  Type'].str.startswith('CRI1-18')]
dfe=dfe[dfe['Injector  Type'].str.startswith('CRI1-16')]
dfe=dfe.select_dtypes(include=['float64'])

#s1=dfe[dfe.ID==1]
#dftemp=pd.DataFrame({"A":[1,2,3,4],"B":['a','b','c','d']})
#dd=pd.concat([s1,dftemp],axis=0,sort=False)
#dd=s1.append(dftemp,sort=False)
#dd.fillna(method='ffill',inplace=True)
#dd.dropna(axis=0,how='any',inplace=True)
#dd.reset_index()
af=pd.DataFrame()
for i in sorted(dfe['ID']):
    if i not in [112,181,182,183,253,257,258] :
        try:
            
            t=getdata(i)
            #print(t)
            #assert(False)
            #t=t[t["ET"]<1000]
            #t.loc[t.InjQ<0,'InjQ']=0
            leftf=dfe[dfe['ID']==i] # prepare left dataframe
            #rightf=t #right dataframe
            joined=leftf.append(t,sort=False)
            joined.fillna(method='ffill',inplace=True)
            joined.dropna(axis=0,how='any',inplace=True)
            af=af.append(joined,sort=False)
            #print("shape of af: ",af.shape)
            #print(af.head(2))
        except Exception as e:
            print(e, " Warning! For test number {0} ".format(i),)

af.to_csv("../x.csv")
af.head(5)
```
