```python
import numpy as np
import matplotlib.pyplot as plt
import math
%matplotlib inline
plt.style.use('seaborn')

def sigmoid(x):
    try:
        return 1/(1+math.exp(-x))
    except OverflowError:
        if x>0:
            return 1/(1+math.exp(-float('inf')))
        else:
            return 1/(1+math.exp(float('inf')))
x1=np.linspace(1,6)

y1=2*x1-2
x2=np.linspace(12,17)
y2=-x2+22
x3=np.linspace(6,12)
y3=x3*0+10
plt.plot(x1,y1)
plt.plot(x2,y2)
plt.plot(x3,y3)
plt.xlabel("x")
plt.ylabel("y")
plt.show()

sx=np.linspace(-10,10)
sy=[sigmoid(i) for i in sx]
plt.plot(sx,sy)
plt.title(r"$\sigma(x)$")
plt.show()

sx=np.linspace(0,20)
#s1=sigmoid(1000i+6000)
sy=[sigmoid(-1000*sigmoid(-1000*i+6000)-1000*sigmoid(1000*i-12000)+500) for i in sx]
plt.plot(sx,sy)
plt.xticks(np.arange(0,20,2))
plt.xlim([0,18])
plt.title(r"$\sigma(-1000*s1-1000*s2+500)$")
plt.show()
```

```python
import tensorflow as tf
mnist = tf.keras.datasets.mnist

(x_train, y_train),(x_test, y_test) = mnist.load_data()
x_train, x_test = x_train / 255.0, x_test / 255.0

model = tf.keras.models.Sequential([
  tf.keras.layers.Flatten(),
  tf.keras.layers.Dense(512, activation=tf.nn.relu),
  tf.keras.layers.Dropout(0.4),
  tf.keras.layers.Dense(10, activation=tf.nn.softmax)
])
model.compile(optimizer='adam',
              batch_size=1024,
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

model.fit(x_train, y_train, epochs=10)
model.evaluate(x_test, y_test)
```

```python
import numpy as np
X = 2 * np.random.rand(100, 1)
y = 4 + 3 * X + np.random.randn(100, 1)
#plt.scatter(X,y)
X_b = np.c_[np.ones((100, 1)), X]

eta = 0.01 # 学习率
n_iterations = 1000
m = 100
theta = np.random.randn(2,1) # 随机初始值
athe=[]
for iteration in range(n_iterations):
    gradients = 2/m * X_b.T.dot(X_b.dot(theta) - y)
    theta = theta - eta * gradients
    athe.append(theta)
plt.plot(np.array(athe)[:,[0]].reshape(-1,1))
plt.plot(np.array(athe)[:,[1]].reshape(-1,1))
theta
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import math
from IPython.core.display import display, HTML
%matplotlib inline
df=pd.read_csv('test.txt',sep='\t',index_col=0)
df=df[df.index<=1500]
df.plot(figsize=(10,8),title='Orignial Q MAP')
df=df.rolling(window=5,center=False).mean()
def ext(col):
    c=len(col)
    last=col.iat[-1]
    lastkey=col.index[-1]
    mid=math.ceil(c*0.8)
    start=col.iat[mid]
    startkey=col.index[mid]
    g=(last-start)/(lastkey-startkey)
    return last+g*(10000-lastkey)
   

df.loc[10000]=df.apply(ext)

df.plot(figsize=(10,8),title="Rolling QMAP mean & extrpolation",xlim=(0,4000),ylim=(0,500))
qlist=[0,0.5,0.75,1.,2.,3.,4.,6.,8.,10.,20.,40.,60.,80.,100.,120.]

plt.style.use('seaborn')
def convert(col):
    col=col.sort_index(ascending=False)
    res=[]
    
    #print(col.name)
    for q in qlist:
        temp=None #store k temperary
        ET=None
        for k,v in col.iteritems():
            #k=float(k)
            if q==v:
                ET=k
                break            
            elif q<v:
                temp=v
                key=k
                continue
            elif temp!=None:
                g=(temp-v)/(key-k)
                
                intq=(g*(key)-temp+q)/g
                ET=intq
                break
        res.append(4000 if ET==None else ET)
    return pd.Series(res,index=qlist,name=col.name)
        

t=[]
t.append(df.apply(convert))
df1=pd.concat(t,axis=1)
ext=pd.Series(2*df1.loc[:,'250000']-df1.loc[:,'300000'],index=qlist,name='100000')

df1=pd.concat([ext,df1],axis=1)
df1.loc[df1.index==0]=0
#df1.columns=df1.columns.str.extract('(\d+)').astype(int).sort_values()

#df1[df1>4000]=4000
#df1.columns.str.extract('(\d+)')

###后期处理，调整非单调区域
for row in range(len(df1.index)):
    for col in range(len(df1.columns)-1)[::-1]:
        if df1.iloc[row,col]<df1.iloc[row,col+1]:
            df1.iloc[row,col]=df1.iloc[row,col+1]
#df1.plot(figsize=(10,8),title="DCM")
#print(df1.to_html())

def myformat(c):
    '''
    format output string as 8.3
    '''
    return "{0:8.3f}".format(c)

with open('x.DCM', 'w') as f:
    head="""
KONSERVIERUNG_FORMAT 2.0
KENNFELD InjVlv_tiET_MAP {0} {1}
 LANGNAME ""
EINHEIT_X "hPa"
EINHEIT_Y "mm^3/inj"
EINHEIT_W "us"
"""
    H,W=df1.shape
    f.write(head.format(W,H))
    f.write('ST/X  '+" ".join(df1.columns.tolist())+"\n")
    for i,row in df1.iterrows():
        f.write("ST/Y {0}\n".format(i))
        f.write("WERT {0}\n".format(" ".join(row.apply(myformat))))
    f.write('END\n')
    #f.write('Hello, world!')
print('DCM has been successfully generated!')


print(df1.to_html())
display(HTML("<H2>Hello world</H2><input type=button value=\"Save DCM\" onclick='alert(\"OK\")'>"))
```
